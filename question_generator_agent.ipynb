```python
from phi.agent import Agent
from phi.model.groq import Groq
from dotenv import load_dotenv
from phi.tools.file import FileTools
from phi.tools.googlesearch import GoogleSearch
from phi.model.openai import OpenAIChat
import os
import base64
load_dotenv()
```

```python
research_agent = Agent(
    # model=Groq(id="llama-3.3-70b-versatile"),
    model=OpenAIChat(id="gpt-4o"),
    tools=[GoogleSearch()], 
    show_tool_calls=True,
    instructions="""
You are a Market Research AI. Given a job role description, your task is to:
1. Search the web (Reddit, Glassdoor, Quora, Blogs) for insights into what makes a top candidate for that role.
2. Summarize in bullet points:
   - Traits of top performers
   - Common mistakes or red flags
   - Differentiators between average vs great candidates
Focus on soft skills, behaviors, and patterns.
""",
)
```

```python
# example role description
role_description = """At Cadence, we hire and develop leaders and innovators who want to make an impact on the world of technology.

The circuit simulation intern will assist R&D staff in developing new algorithms and software for the Spectre circuit simulator. Job duties may include hands-on experience in researching new algorithmic approaches, software engineering to implement new features, performing software validation and testing, and investigating performance/accuracy issues in the simulator.

Primary responsibilities include researching and developing Machine Learning approaches to problems in the EDA and system design. Applicant must possess a working knowledge of various Machine Learning techniques such as linear regression, classification, clustering, decision trees, and neural networks, and have a practical experience with ML libraries, such as Sklearn and PyTorch. Solid understanding of statistics is a plus.

Applicant should be a candidate for an advanced degree (Masters or PhD) in the electrical engineering, computer science, applied mathematics, or related field.

We’re doing work that matters. Help us solve what others can’t."""
```

```python
market_research = research_agent.run(f"Do market research for this role: {role_description}")
print(market_research)
```

```python
question_generator = Agent(
    name="question_generator",
    llm=OpenAIChat(id="gpt-4o"),
    instructions="""
You are an expert interviewer AI.

Your goal is to generate 5 **non-technical, high-signal** interview questions that help hiring managers identify **top 10% candidates** for a given role. These questions should:
- Be answerable in ~2 minutes
- Reveal behaviors, decision-making, and mindset
- Avoid clichés or generic soft-skill questions

Given a role description and market research summary, craft questions that are sharp, specific, and interesting. Include a short **note** under each question on what kind of answer signals a strong candidate.

Format:
1. Question
   - Note: What a great answer should include
""",
)
```

```python
response = question_generator.run(
    f"""
Role Description: {role_description}
Market Research Summary: {market_research}
"""
)

print(response)
```
